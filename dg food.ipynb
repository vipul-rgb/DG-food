# Data Normalization 

```python
from sklearn.preprocessing import normalize 

dt1['country_mean'] = pd.Series(dt.mean(axis = 1), index = dt1.index)
country_mean_max = dt1['country_mean'].max()
country_mean_min = dt1['country_mean'].min()

country_mean_scaling = (dt1.country_mean-country_mean_min) / country_mean_max
dt1['country_mean_scaling'] = pd.Series(country_mean_scaling, index = dt1.index)
dt1.head()
```

```python
dt1['country_sum'] = pd.Series(dt.sum(axis=1), index=dt1.index)
country_sum_max = dt1['country_sum'].max()
country_sum_min = dt1['country_sum'].min()
country_sum_scaled = (dt1.country_sum-country_sum_min) / country_sum_max
dt1['country_sum_scaled'] = pd.Series(country_sum_scaled, index=dt1.index)
dt1.head()
```

```python
dt1.plot(kind='scatter', x='Pc2', y='Pc1', s=dt1['country_sum_scaled']*100, figsize=(16,8))
plt.show()
```

```python
dt1['country_change'] = pd.Series(dt['2007']-dt['1990'], index=dt1.index)
country_change_max = dt1['country_change'].max()
country_change_min = dt1['country_change'].min()
country_change_scaled = (dt1.country_change - country_change_min) / country_change_max
dt1['country_change_scaled'] = pd.Series(country_change_scaled, index=dt1.index)
dt1[['country_change','country_change_scaled']].head()
```

```python
dt1.plot(kind='scatter', x='Pc2', y='Pc1', s=dt1['country_change_scaled']*100, figsize=(16,8))
```

# Algorithms 

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit(dt)
```

```python
dt1['cluster'] = pd.Series(clusters.labels_, index=dt1.index)
dt1.head()
```

```python
# plot the data 
#sns.set_style('whitegrid')
sns.lmplot(data=dt1, x="Pc2", y="Pc1" , hue = 'cluster')
plt.show()
```

```python
scatter_plot = dt1.plot(kind ='scatter', x ='Pc2',y='Pc1' , figsize=(15,7))
for i , Country in enumerate(dt.index):
    scatter_plot.annotate(Country ,(dt1.iloc[i].Pc2 ,dt1.iloc[i].Pc1 ))
```

```python
sns.countplot(dt1, x="cluster")
plt.show()
```

```python
sse = []
k_rng = range(1,10)

for k in k_rng:
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(dt1[['Pc2','Pc1']])
    sse.append(kmeans.inertia_)
```

```python
plt.xlabel('k')
plt.ylabel('sum of squared error')
plt.plot(k_rng,sse)
plt.show()
```
